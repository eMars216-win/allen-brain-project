```python
import requests

url = (
    "http://api.brain-map.org/api/v2/data/query.json"
    "?criteria=model::Structure"
    ",rma::criteria,ontology[name$eq'Human+Brain+Atlas']"
    "&rma::options=[only$eq'id,name,acronym,parent_structure_id,hemisphere_id,graph_order']"
    "&num_rows=1839"
)

response = requests.get(url, timeout=30)
data = response.json()
structures = data['msg']

# Look at the thalamus parent (4432) specifically
# to see if graph_order reveals the pairing pattern
thalamus_children = [s for s in structures if s['parent_structure_id'] == 4432]
thalamus_children.sort(key=lambda x: x['graph_order'])

print("Thalamus children sorted by graph_order:")
for s in thalamus_children:
    hemi = {1:'LEFT', 2:'RIGHT', 3:'BILATERAL'}.get(s['hemisphere_id'], '?')
    print(f"  graph_order={s['graph_order']:4d} | {hemi:9s} | id={s['id']:8d} | {s['name']}")

# Also check the posterior cortical nucleus (4369)
print()
pcn_children = [s for s in structures if s['parent_structure_id'] == 4369]
pcn_children.sort(key=lambda x: x['graph_order'])
print("Posterior cortical nucleus children sorted by graph_order:")
for s in pcn_children:
    hemi = {1:'LEFT', 2:'RIGHT', 3:'BILATERAL'}.get(s['hemisphere_id'], '?')
    print(f"  graph_order={s['graph_order']:4d} | {hemi:9s} | id={s['id']:8d} | {s['name']}")
```

```python
def build_hemisphere_pairs(structures_df):
    """
    Build left/right pairs using hemisphere_id field and parent_structure_id.
    
    Confirmed from API exploration:
    - hemisphere_id 1=left, 2=right, 3=bilateral
    - 313 parent structures have equal left and right child counts
    - Zero mismatches — positional pairing by graph_order is safe
    """
    from collections import defaultdict
    import pandas as pd

    structures = structures_df.to_dict('records')
    
    left_by_parent  = defaultdict(list)
    right_by_parent = defaultdict(list)

    for s in structures:
        if s['hemisphere_id'] == 1:
            left_by_parent[s['parent_structure_id']].append(s)
        elif s['hemisphere_id'] == 2:
            right_by_parent[s['parent_structure_id']].append(s)

    shared_parents = set(left_by_parent.keys()) & set(right_by_parent.keys())

    pairs = []
    for parent_id in shared_parents:
        left_children  = sorted(left_by_parent[parent_id],  key=lambda x: x['graph_order'])
        right_children = sorted(right_by_parent[parent_id], key=lambda x: x['graph_order'])

        for left, right in zip(left_children, right_children):
            pairs.append({
                'parent_structure_id': parent_id,
                'left_id':            left['id'],
                'left_name':          left['name'],
                'left_acronym':       left['acronym'],
                'right_id':           right['id'],
                'right_name':         right['name'],
                'right_acronym':      right['acronym'],
            })

    df = pd.DataFrame(pairs)
    log.info(f"Built {len(df)} hemisphere pairs from {len(shared_parents)} shared parents")
    return df
```
